```python
import requests
import pycandela

# Fetch the data
response = requests.get('https://raw.githubusercontent.com/vega/vega-datasets/gh-pages/data/iris.json')

# Check the response content
print(response.text)

# Parse the data as JSON
data = response.json()

# Use Candela to create a scatter plot
pycandela.components.ScatterPlot(
    data=data, color='species', x='sepalLength', y='sepalWidth'
)
```
