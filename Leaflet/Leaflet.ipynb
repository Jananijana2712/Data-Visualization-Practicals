```python
import folium
import pandas as pd
```

```python
# Sample data for demonstration purposes
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Latitude': [40.7128, 34.0522, 41.8781, 29.7604, 33.4484],
    'Longitude': [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740],
    'Value': [100, 150, 200, 250, 300]
}
```

```python
# Convert the data to a DataFrame
df = pd.DataFrame(data)
```

```python
# Create maps with titles for each location
maps = []
for index, row in df.iterrows():
    m = folium.Map(location=[row['Latitude'], row['Longitude']], zoom_start=10)
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=f"{row['City']}: Value {row['Value']}",
        tooltip=row['City']
    ).add_to(m)
    
    title_html = f'''
    <h3 align="center" style="font-size:16px;"><b>Map of {row['City']}</b></h3>
    '''
    m.get_root().html.add_child(folium.Element(title_html))
    
    maps.append(m)
```

```python
# Display all maps
for m in maps:
    display(m)
```
